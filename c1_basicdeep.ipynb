```python
#Workshop1 import iris's information

import seaborn as sns
import matplotlib.pyplot  as plt

#โหลดข้อมูล
df = sns.load_dataset('iris')
#ข้อมูลตัวอย่าง5แถว
df.head()
```

```python
#เช็คข้อมูลหาย
df.isnull().sum()  
```

```python
#พล็อตกราฟ

sns.scatterplot('sepal_length','sepal_width',data=df , hue='species')
plt.show()
```

```python
sns.scatterplot('petal_length','petal_width',data=df , hue='species')
plt.show()
```

```python
sns.pairplot(df,hue='species',height=2)
plt.show()
```

```python
#Workshop2  prapare informaton such as feature and Lable
X = df.drop('species',axis=1)
```

```python
feature_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
X = df[feature_cols]
X.head()
```

```python
y = df.species
y.head()
```

```python
X.shape
y.shape
```

```python
#Workshop 3 Train and Test 

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score 


#แบ่ง train 30%
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size=0.3) 
```

```python
#ตรวจสอบชนิดของแต่ละสายพันธุ์
import numpy as np 
(lable,count)=np.unique(y_test,return_counts=True)
freq = np.asanyarray((lable,count)).T
print(freq)
```

```python
#Workshop 4  Create Model and Testing 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# Train
from sklearn.svm import SVC
model= SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
#Test 
y_pred = model.predict(X_test)
print('Accuracy Score: {:.4f}'.format(accuracy_score(y_test,y_pred)))
print(classification_report(y_test,y_pred))
cm = confusion_matrix(y_test,y_pred)
print(cm)
```
