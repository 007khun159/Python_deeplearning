```python
#Doload infor about cancer 

from sklearn.datasets import load_breast_cancer
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

cancer = load_breast_cancer()
cancer.keys()  #check key infor
```

```python
# ข้อมูล and ชื่อตัวแปร  (feture) ตัดเอามาเพียง  2 colum

df = pd.DataFrame(cancer.data[:,0:2],columns=cancer.feature_names[0:2])
df['class'] = cancer.target   #class/target/label(มะเร็ง หรือ ไม่เป็น)
df.sample(5,random_state=1)
```

```python
#change class 0 1 to String
#เปลี่ยน calss เป็น class ด้วย
#df['class']=np.where(df['calss']==0, 'yes','no')
#df['class']=np.where(df['calss']==0, 'maliganant','begin')

target_map = {0:'malignant',1:'begin'}   #chang 0 1 to string 
df['calss'] = df['class'].map(target_map)

df.sample(5,random_state=1)
```

```python
#check loss infor 

df.isnull().sum()
```

```python
#Plot graph 

plt.figure(figsize=(5.2, 4))

sns.scatterplot(x='mean radius', y='mean texture', data=df, hue='class', style='class', 
                alpha=0.9, edgecolor='w', s=80)

plt.show()
```
